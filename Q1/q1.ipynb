```python
import csv
import matplotlib.pyplot as plt

def main():
    f = []
    f.append(open('every.csv', 'r', encoding = 'cp949'))
    f.append(open('seoul.csv', 'r', encoding = 'cp949'))
    f.append(open('daejeon.csv', 'r', encoding = 'cp949'))
    f.append(open('busan.csv', 'r', encoding = 'cp949'))
    f.append(open('jeju.csv', 'r', encoding = 'cp949'))
    data = []
    for i in range(0, len(f)):
        data.append(csv.reader(f[i], delimiter = ','))
        next(data[i])
    
    avg = [[], [], [], [], []]
    for i in range(0, len(data)):
        for row in data[i]:
            month = int(row[0].split('-')[1])
            avg[i].append(float(row[2]))
        
    plt.title("Monthly Average Temperature")
    plt.xlabel("Month")
    plt.xticks(range(0, 12), labels = range(1, 13))
    plt.ylabel("Avg Temperature")
    plt.plot(avg[0], label = "Every")
    plt.plot(avg[1], label = "Seoul")
    plt.plot(avg[2], label = "Daejeon")
    plt.plot(avg[3], label = "Busan")
    plt.plot(avg[4], label = "Jeju")
    plt.legend()
    plt.show()

    for i in range(0, len(f)):
        f[i].close()
        
if __name__ == '__main__':
    main()
```
