```python
import csv
import matplotlib.pyplot as plt

def main():
    f = open('jeju.csv', 'r', encoding = 'cp949')
    data = csv.reader(f, delimiter = ',')
    next(data)
    fig, ax = plt.subplots(2, 2)
    fig.set_size_inches(10, 10)
    axes = [ax[0, 0], ax[0, 1], ax[1, 0], ax[1, 1]]
    axes[0].set_title('Jeju\'s Population in 2010')
    axes[1].set_title('Jeju\'s Population in 2014')
    axes[2].set_title('Jeju\'s Population in 2018')
    axes[3].set_title('Jeju\'s Population in 2022')
    for a in axes:
        a.set_ylabel('Population(%)')
    xdata = ['Man', 'Woman']
    i = 0
    for row in data:
        row.remove(row[0])
        for j in range(0, len(row)):
            row[j] = int(row[j])
        result = []
        result.append(int((row[0] / (row[0] + row[1])) * 10000) / 100) #전체에서 남자 인구수의 비율을 소수점 둘째 자리까지 저장
        result.append(int((row[1] / (row[0] + row[1])) * 10000) / 100) #전체에서 여자 인구수의 비율을 소수점 둘째 자리까지 저장
        bar = axes[i].bar(xdata, result, color = ['b', 'r'], alpha = 0.5)
        for rect in bar:
            height = rect.get_height()
            axes[i].text(rect.get_x() + rect.get_width() / 2, height + 0.1, height, ha = 'center', va = 'bottom')
        i += 1
    plt.show()
    f.close()
    
    
if __name__ == '__main__':
    main()
```
