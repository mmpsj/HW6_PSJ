```python
import csv
import matplotlib.pyplot as plt
import random

def makeRandom(writer, n):
    result = [0, 0, 0, 0, 0, 0]
    for i in range(0, n):
        result[random.randrange(1, 7) - 1] += 1
    writer.writerow([n, result[0], result[1], result[2], result[3], result[4], result[5]])

def main():
    f = open('dice.csv', 'w', encoding = 'cp949')
    f.close()
    f = open('dice.csv', 'a', encoding = 'cp949', newline = '')
    writer = csv.writer(f)
    
    writer.writerow(['시행 횟수', 1, 2, 3, 4, 5, 6])
    makeRandom(writer, 100)
    makeRandom(writer, 1000)
    makeRandom(writer, 10000)
    makeRandom(writer, 100000)
    
    f.close()
    
    f = open('dice.csv', 'r', encoding = 'cp949')
    data = csv.reader(f, delimiter = ',')
    header = next(data)
    fig, ax = plt.subplots(2, 2)
    fig.set_size_inches(10, 10)
    plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
    axes = [ax[0, 0], ax[0, 1], ax[1, 0], ax[1, 1]]
    axes[0].set_title('Random Dice Result (100 times)')
    axes[1].set_title('Random Dice Result (1000 times)')
    axes[2].set_title('Random Dice Result (10000 times)')
    axes[3].set_title('Random Dice Result (100000 times)')
    for i in range(0, len(axes)):
        axes[i].set_xlabel('Dice Number')
        axes[i].set_ylabel('Count')
    temp = 0
    xdata = [1, 2, 3, 4, 5, 6]
    for row in data:
        result = []
        for i in range(1, len(row)):
            for j in range(0, int(row[i])):
                result.append(i)
        n, bins, patches = axes[temp].hist(result, bins = 6)
        for rect in patches:
            axes[temp].text(rect.get_x() + rect.get_width() / 2, rect.get_height() + 0.1, int(rect.get_height()), ha = 'center', va = 'bottom')
        temp += 1
    plt.show()
    f.close()
    
if __name__ == '__main__':
    main()
```
