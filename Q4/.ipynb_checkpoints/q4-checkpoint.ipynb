```python
import csv
import matplotlib.pyplot as plt

def main():
    f = open('202303_Subway.csv', 'r', encoding = 'cp949')
    data = csv.reader(f, delimiter = ',')
    next(data)
    next(data)
    get = [{}, {}, {}] #순서대로 승차, 하차, 승하차에 대한 딕셔너리. 여러 호선이 지나가는 역은 데이터가 여러 개 있으므로 딕셔너리로 구현함.
    for row in data:
        if row[1] in get[0].keys():
            get[0][row[1]] += int(row[2]) + int(row[4])
            get[1][row[1]] += int(row[3]) + int(row[5])
            get[2][row[1]] += int(row[2]) + int(row[3]) + int(row[4]) + int(row[5])
        else:
            get[0][row[1]] = int(row[2]) + int(row[4])
            get[1][row[1]] = int(row[3]) + int(row[5])
            get[2][row[1]] = int(row[2]) + int(row[3]) + int(row[4]) + int(row[5])
    for i in range(0, len(get)):
        get[i] = sorted(get[i].items(), key=lambda x: x[1], reverse=True)
    fig, ax = plt.subplots(3, 1)
    fig.set_size_inches(25, 20)
    plt.rc('font', family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus'] = False
    plt.subplots_adjust(hspace = 0.5)
    #axes = [ax[0], ax[1], ax[2]]
    ax[0].set_title('출근시간 승차 승객 수')
    ax[1].set_title('출근시간 하차 승객 수')
    ax[2].set_title('출근시간 승하차 승객 수')
    for i in range(0, len(get)):
        name = []
        count = []
        for j in range(0, 30):
            name.append(get[i][j][0])
            count.append(get[i][j][1])
        ax[i].bar(name, count)
        ax[i].tick_params(axis = 'x', rotation=45)
    plt.show()
    f.close()
    
if __name__ == '__main__':
    main()
```
